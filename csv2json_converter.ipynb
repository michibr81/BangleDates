```python
import vobject
import datetime as dt
import tzlocal
import pandas as pd

removeStrings = ["Abfuhrtermin"]    #strings to remove from event-summary

loctz = tzlocal.get_localzone() #local timezone
data = open("../personal.ics").read()   # read dates from ics-file

df = pd.DataFrame()

for cal in vobject.readComponents(data):
    for component in cal.components():
        if component.name == "VEVENT":
            start = component.dtstart.valueRepr()

            isOnlyDate = type(start) == dt.date
            if isOnlyDate:
                start = dt.datetime(start.year,start.month,start.day,tzinfo=loctz)
                
            if start > dt.datetime.now(loctz):

                eventSummary = component.summary.valueRepr()

                for removeString in removeStrings:
                    eventSummary = eventSummary.replace(removeString,"")

                df = df.append({"Event":eventSummary, "Date":str(start.date()), "AllDay":isOnlyDate},ignore_index=True)
                
df.to_json("../personal.json",orient="records")
df
```
